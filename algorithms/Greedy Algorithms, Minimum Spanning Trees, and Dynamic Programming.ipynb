## Minimum spanning tree (MST) problem

Minimum spanning tree problem: connection bunch all points together in minimum distance

Consider links between points as undirected graph of Vertices (V, points) and edges, E

O(n log n) algorithms given below. More complicated randomised Karger-Klein-Tarjan algorithm can do it in O(n) time. Or deterministic Chazelle algorithm O(log* n), where log* means the log-of-the-log-of-the-log(etc until n goes below 1, then that's your value - this is very close to linear time)




Deterministic algorithm: no randomisation so gives the same result on every run if input is constant



## Prim's algorithm for MST

Similar to Djikstra

Start from a point, then traverse, always making the next move from the point with the lowest distance (is a greedy algo)

Says Prim's algo 100% always makes the optimal tree

Prim's algo runs in O(V * E) time, or you can use a heap to store whichever edge or node comes next, letting it run in O(n log n) time




## Kruskal's algorithm for MST

Like Prim's, but instead of traversing from a point, just 'traverse' each edge in order of ascending distance (aka cost). So you're 'traversing' edges which aren't necessarily connected.

Run it until all nodes are connected. 

Use union-find to merge all visited nodes into that 'group' (aka, in this case, 'component'). Presumably so you know it's been visited

Gets you optimal tree 100% of the time.

Union operation: merge two groups in the structure





## Minimum spanning trees applied to clustering

Unsupervised clustering into k clusters

Function to maximise: spacing between clusters

Greedy approach: treat all points as their own 'cluster', then find smallest gap between points and 'merge' those two so you have one less cluster, repeat this greedily until you have k clusters 

Can think of this as Kruskal's algorithm, but stopping early when there are k clusters left

This is an example of Single Link Clustering: where the algo works by pairs nodes, one at a time

A drawback of Single Link Clustering: it tends to produce long thin clusters in which nearby elements of the same cluster have small distances, but elements at opposite ends of a cluster may be much farther from each other than two elements of other clusters. 





## Huffman codes

Want to make optimal encoding of an alphabet (or anything) in bytes. Might do this by giving most frequent values fewer bytes. Eg: 'e' = '0', however this would require all other codes to start their binary sequences with '0', so they end up taking up more space: the best combination of the two is something one can search for with an algo, of which Huffman's algo is one.

Can think of the bits to chose as decision nodes on a binary decision tree, the leaves of which correspond to letters. 

The algo makes said decision tree using a greedy bottom up approach:

>Take the least frequent letter in alphabet and merge it with 2nd least frequent, making a first node at the base of the tree

>Of the remaining letters and the new merged node (which has a frequency equal to the sum of it's component letters), repeat the above step to make a new node. Continue until all letters are done.

Not sure if the above is "Huffman algorithm" but it's the one described in the lecture 






## Dynamic programming

Dynamic Programming: algorithmic technique for solving a problem by recursively breaking it down into simpler subproblems. Then using the fact that the optimal solution to the overall problem depends upon the optimal solution to it's individual subproblems.

Might involve memoising subproblems so each is only solved once.

Requirements of DP:

>it's possible to identity a small number of subproblems

>solving the smaller subproblems makes solving the bigger subproblems much faster

Sometimes the overall problem is the "biggest" subproblem

The key is to pick good subproblems
