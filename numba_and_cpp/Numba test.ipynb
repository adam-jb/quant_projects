### Quicklist to jit-ify a piece of code:

add @njit
don't use enumerate; use for i in range(len(thing)):.. etc..
numpy inputs not lists




### If you see a warning about 'reflected list' it probs mean you're using a list when you should be using a numpy array

```python
%pylab inline



from numba import jit, njit, vectorize
import random
from concurrent.futures import ThreadPoolExecutor



def monte_carlo_pi(nsamples):
    acc = 0
    for i in range(nsamples):
        x = random.random()
        y = random.random()
        if (x ** 2 + y ** 2) < 1.0:
            acc += 1
    return 4.0 * acc / nsamples
```

```python
%time monte_carlo_pi(1000000)   # wall time = actual time from start to end of program
```

```python
monte_carlo_pi_jit = jit()(monte_carlo_pi)  # adding jit
```

```python
%time monte_carlo_pi_jit(1000000)        # takes longer the first time run as compiling: x20 faster after that
```

```python
# example of doing something wrong: numba (unlike python lists) doesnt do at all well with mixed
# types: treat them more like numpy arrays

# also: use njit() instead of jit(), as njit() wont let you use things like mixed types, which
# slow you down

def original_function(item_list):
    output_list = []
    for item in item_list:
        if item % 2 == 0:
            output_list.append(2)
        else:
            output_list.append(1)  
    return output_list

test_array = list(range(100000))
test_array = np.arange(1000000)

tt = %time original_function(test_array)
print(tt)
original_function_jit = jit()(original_function)
tt = %time original_function_jit(test_array)
print(tt)
tt = %time original_function_jit(test_array)
print(tt)
```

```python
def original_function(item_list):
    output_list = []
    for item in item_list:
        if item % 2 == 0:
            output_list.append(2)
        else:
            output_list.append(1)  
    return output_list
```

```python
%time original_function(test_array)  # if timing in jupyter, give operation it's own cell
```

```python
# it's faster!
original_function_jit = njit(nogil=True)(original_function) # nogil doesnt seem to help tho (even with
                    # threadExecutor, which you also need set up to get the most from this)
```

```python
%time original_function_jit(test_array)
```

```python
# about the same as non-jitted above. Not blown away yet
@vectorize
def scalar_computation(num):
    if num % 2 == 0:
        return 2
    else:
        return 1
    
```

```python
%time scalar_computation(test_array)
```
